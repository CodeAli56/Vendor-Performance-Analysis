
###### The purchases table contains actual purchase data, including the date of purchase, product (brands) purchased by vendors, the amount paid and the quantity purchased.
###### The purchase price column is derived from purchase_price table, which provide product-wise actual and purchase price. The combination of vendor and brand is unique.
###### The vendor_invoice table aggregates data from purchases table, summarizing quantity and amount, along with additional columns like freight. The table maintains uniquesness based on vendor and PO number.
###### The sales table captures actual sales transaction, detailing the brand purchased by vendors, the quantity sold, the selling price, and the revenue earned.

###### As the data we need for analysis is distributed in different tables, we need to create a summary table containing:
###### 1. purchase transactions made by vendors
###### 2. Sales transaction data
###### 3. freight cost for each vendor
###### 4. actual product price from vendors

```python
vendor_invoice = pd.read_csv("data/vendor_invoice.csv")
sales = pd.read_csv("data/sales.csv")
purchases = pd.read_csv("data/purchases.csv")
purchase_prices = pd.read_csv("data/purchase_prices.csv")
```

```python
prices_summary = pd.read_sql(
    """ Select
            p.vendorNumber, 
            p.vendorName, 
            p.Brand,
            p.Description,
            p.PurchasePrice,
            pp.Volume,
            pp.Price as ActualPrice,
            sum(p.Quantity) as TotalPurchasedQuantity,
            sum(p.Dollars) as TotalPurchasedDollar
        from purchases p join purchase_prices pp on p.Brand = pp.Brand
        where p.purchasePrice > 0
        group by 1,2,3      
        order by 8                    
    """, conn)

prices_summary
```

```python
sales.columns
```

```python
sales_summary = pd.read_sql("""
    select
        vendorNo as VendorNumber,
        Brand,
        round(sum(SalesDollars), 2) as TotalSalesDollars,
        round(sum(SalesPrice), 2) as TotalSalesPrice,
        round(sum(SalesQuantity), 2) as TotalSalesQuantity,
        round(sum(ExciseTax),2 ) as TotalExcisetax           
    from sales
    group by 1, 2
    order by 3
""", conn)

sales_summary
```

```python
priceSalesSummary = pd.merge(prices_summary, sales_summary, left_on=["VendorNumber", "Brand"], right_on=["VendorNumber", "Brand"], how='left')
```

```python
vendor_summary = pd.merge(priceSalesSummary, freight_summary, on="VendorNumber", how='left')
```

###### Here, vendor_summary and vendor_sales_summary are same, just 2 different approaches to summarize all 3 tables into 1. We will use Vendor Summary.

This query generates a vendor-wise sales and purchase summary.
1. This query resolves heavy joins and aggregations on large dataset like purchase and sales.
2. Storing this resultant dataset for faster Dashboarding and reporting.

### Cleaning Vendor_summary data

```python
vendor_summary.dtypes
```

```python
vendor_summary.isnull().sum()       
# here, 178 means there are such cases where vendor couldn't sell those products which  he bought earlier.
```

```python
vendor_summary['Volume'] = vendor_summary['Volume'].astype("float")
```

```python
vendor_summary['VendorName'].unique()
```

```python
vendor_summary["VendorName"] = vendor_summary["VendorName"].apply(lambda x: x.strip())
```

```python
vendor_summary['VendorName'].unique()
```

```python
vendor_summary.fillna(value=0, inplace=True)
```

```python
vendor_summary
```

Creating new columns for better analysis.

```python
vendor_summary["GrossProfit"] = (vendor_summary["TotalSalesDollars"] - vendor_summary['TotalPurchasedDollar']).round(2)
```

```python
vendor_summary
```

```python
vendor_summary['ProfitMargin'] = (vendor_summary['GrossProfit']/vendor_summary['TotalSalesDollars']*100).round(2)
```

```python
vendor_summary['StockTurnOver'] = vendor_summary['TotalSalesQuantity'] / vendor_summary['TotalPurchasedQuantity']
```

```python
vendor_summary['SalesToPurchaseRatio'] = (vendor_summary['TotalSalesDollars'] / vendor_summary["TotalPurchasedDollar"]).round(2)
```

#### Now, creating a table to store the cleaned dataset into the database.

```python
cursor = conn.cursor()
```

```python
vendor_summary.columns
```

```python
cursor.execute("""
    create table vendor_summary (
        VendorNumber int,
        vendorName varchar(100),
        Brand int,
        Description varchar(100),
        PurchasePrice decimal(10,2),
        ActualPrice decimal(10, 2),
        Volume decimal(10, 2),
        TotalPurchaseQuantity int,
        TotalPurchasedDollar decimal(15,2),
        TotalSalesDollars decimal(15,2),
        TotalSalesPrice decimal(15,2),
        TotalSalesQuantity decimal(15,2),
        TotalExcisetax decimal(15,2),
        FreightCost decimal(15,2),
        GrossProfit decimal(15,2),
        ProfitMargin decimal(15,2),
        StockTurnOver decimal(15,2),
        SalesToPurchaseRatio decimal(15,2),
        PRIMARY KEY (VendorNumber, Brand)
    )
""")
```

```python
pd.read_sql("select * from vendor_summary",conn)        # successfully created table in the database.
```

```python
# Now inserting data in the created table

vendor_summary.to_sql("vendor_summary", conn, if_exists="replace", index=False)
```
